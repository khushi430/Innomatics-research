```python
def analyze_post_engagement(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status


# Example
likes = [200, 300, 250, 300]
total, result = analyze_post_engagement(likes)
print("Total Likes:", total)
print("Post Status:", result)
```

```python
def check_rainfall(rainfall_list, required_level):
    total = 0
    for rain in rainfall_list:
        total += rain

    average = total / len(rainfall_list)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average, status


# Example
rainfall = [70, 80, 75, 65]
avg, status = check_rainfall(rainfall, 70)
print("Average Rainfall:", avg)
print("Rainfall Status:", status)
```

```python
def detect_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        return "Yes"
    else:
        return "No"


# Example
users = ["john", "alice", "john", "mike"]
print("Duplicate Accounts Found:", detect_duplicates(users))
```

```python
def filter_premium_crops(prices):
    premium = []
    for price in prices:
        if price > 2000:
            premium.append(price)
    return premium


# Example
crop_prices = [1500, 2500, 3200, 1800]
print("Premium Crops:", filter_premium_crops(crop_prices))
```

```python
def check_application_health(error_count):
    if error_count == 0:
        return "Healthy"
    elif error_count <= 5:
        return "Minor Issues"
    else:
        return "Critical Issues"


# Example
errors = 7
print("Error Count:", errors)
print("System Status:", check_application_health(errors))
```

```python
def check_exam_eligibility(attendance_list):
    total_classes = len(attendance_list)
    attended = 0

    for status in attendance_list:
        if status == 1:  # 1 means present
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    return percentage, eligibility


# Example
attendance = [1,1,1,0,1,1,1,1,0,1]
percent, result = check_exam_eligibility(attendance)
print("Attendance Percentage:", percent)
print("Exam Eligibility:", result)
```

```python
def calculate_electricity_bill(units):
    bill = 0

    # Billing calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif 500 <= bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


# Example
units = 250
bill_amount, usage_status = calculate_electricity_bill(units)
print("Total Bill:", bill_amount)
print("Usage Status:", usage_status)
```
