```python

# Problem 1: Unique Words in a Sentence
# ------------------------------------------------------------
print("Problem 1: Unique Words in a Sentence")

sentence = "python is easy and python is powerful"

# Split sentence into words
words = sentence.split()

# Convert list to set to remove duplicates
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
print("\n")


# Problem 2: Highest Salary from Employee Data
# ------------------------------------------------------------
print("Problem 2: Highest Salary from Employee Data")

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print("Highest Salary:", highest_employee, "-", highest_salary)
print("\n")



# Problem 3: Find Maximum and Minimum Values
# ------------------------------------------------------------
print("Problem 3: Find Maximum and Minimum Values")

numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through list to compare values
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
print("\n")



# Problem 4: Count Products Above a Price Threshold
# ------------------------------------------------------------
print("Problem 4: Count Products Above 1000")

prices = [450, 1200, 899, 1500, 300]

count = 0

# Count prices greater than 1000
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
print("\n")



# Problem 5: Calculate Attendance Percentage
# ------------------------------------------------------------
print("Problem 5: Calculate Attendance Percentage")

attendance = ["P", "P", "A", "P", "P"]

# Count number of Present days
present_count = attendance.count("P")

# Calculate percentage
percentage = (present_count / len(attendance)) * 100

print("Attendance Percentage:", percentage)
print("\n")



# Problem 6: Remove Duplicate Phone Numbers
# ------------------------------------------------------------
print("Problem 6: Remove Duplicate Phone Numbers")

phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set to remove duplicates
unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
print("\n")



# Problem 7: Count Character Frequency
# ------------------------------------------------------------
print("Problem 7: Count Character Frequency")

text = "python"
frequency = {}

# Count each character occurrence
for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print("Character Frequency:", frequency)
print("\n")



# Problem 8: Convert List to Tuple
# ------------------------------------------------------------
print("Problem 8: Convert List to Tuple")

list_data = [10, 20, 30]

# Convert list to tuple
tuple_data = tuple(list_data)

print("Tuple:", tuple_data)
print("\n")



# Problem 9: Check if a Key Exists in a Dictionary
# ------------------------------------------------------------
print("Problem 9: Check if a Key Exists in a Dictionary")

employees_check = {
    "Ravi": 75000,
    "Anita": 68000
}

key_to_check = "Ravi"

# Check if key exists
if key_to_check in employees_check:
    print("Employee exists")
else:
    print("Employee does not exist")

print("\n")



# Problem 10: Calculate Average Marks
# ------------------------------------------------------------
print("Problem 10: Calculate Average Marks")

marks = [70, 75, 80, 65]

# Calculate average
total_marks = sum(marks)
average_marks = total_marks / len(marks)

print("Average Marks:", average_marks)
```
