```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        # Condition 1: Minimum amount check
        if amount < 50:
            print("Recharge amount must be at least ₹50. Try again.")
            continue

        # Condition 2: Valid plan check
        if amount in valid_plans:
            print("Recharge Successful ✅")
            break
        else:
            print("Invalid plan. Available plans:", valid_plans)
            print("Please try again.\n")

# Call function
validate_recharge()
```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert ⚠")
        else:
            print(f"{product}: Stock OK ✅")

# Sample Data
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 8,
    "Monitor": 20
}

check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System

def process_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return f"Average: {average:.2f} → Pass ✅"
    else:
        return f"Average: {average:.2f} → Fail ❌"

# Example
student_marks = [60, 45, 70, 55]
print(process_result(student_marks))
```

```python
# Problem 4: Cab Fare Estimator

def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12

    fare = base_fare + (distance * per_km)

    if peak:
        fare *= 1.25   # 25% extra

    return fare


while True:
    distance = float(input("Enter distance in km: "))
    peak_input = input("Is it peak hour? (yes/no): ").lower()

    peak = True if peak_input == "yes" else False

    total_fare = calculate_fare(distance, peak)
    print(f"Total Fare: ₹{total_fare:.2f}")

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

```python
# Problem 5: Attendance Eligibility Checker

def check_attendance(attendance_list):
    present_days = 0

    for status in attendance_list:
        if status == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% → Eligible ✅"
    else:
        return f"Attendance: {percentage:.2f}% → Not Eligible ❌"

# Example
attendance = ["P", "A", "P", "P", "A", "P", "P"]
print(check_attendance(attendance))
```

```python
# Problem 6: Password Strength Checker

def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Password Weak ❌ (Minimum 8 characters required)"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Password Strong ✅"
    else:
        return "Password Weak ❌ (Must contain at least one digit and one special character @#$)"

# Example
pwd = input("Enter password: ")
print(check_password(pwd))
```
