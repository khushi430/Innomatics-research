```python
# Problem 1: Smart Parking Lot Management System

def smart_parking_system(capacity, vehicle_logs):
    """
    This function manages a smart parking lot.
    It calculates current parked vehicles,
    identifies peak usage, and checks capacity overflow.
    """

    current_count = 0
    peak_count = 0

    for log in vehicle_logs:
        if log == "IN":
            current_count += 1
        elif log == "OUT":
            current_count -= 1

        # Track peak usage
        if current_count > peak_count:
            peak_count = current_count

    print("Currently Parked Vehicles:", current_count)
    print("Peak Parking Usage:", peak_count)

    if current_count > capacity:
        print("Parking Status: Exceeded Capacity ðŸš¨")
    else:
        print("Parking Status: Available âœ…")


# Sample Input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, logs)
```

```python
# Problem 2: Online Food Delivery Time Estimator

def estimate_delivery_time(distance, traffic_level, weather):
    """
    This function estimates delivery time based on:
    Distance, Traffic level, and Weather condition.
    """

    base_time = distance * 5   # Assume 5 minutes per km

    # Traffic adjustment
    if traffic_level.lower() == "high":
        base_time += 15
    elif traffic_level.lower() == "medium":
        base_time += 10
    else:
        base_time += 5

    # Weather adjustment
    if weather.lower() == "rainy":
        base_time += 10
    elif weather.lower() == "stormy":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")


# Sample Input
estimate_delivery_time(8, "High", "Rainy")
```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer

def seat_occupancy(total_seats, booked_seats):
    """
    This function calculates occupancy percentage
    and determines show status.
    """

    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", round(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull ðŸŽ¬")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Sample Input (150 booked seats)
seat_occupancy(200, [1]*150)
```

```python
# Problem 4: Cloud Server Load Classification System

def classify_server_load(cpu_readings):
    """
    This function calculates average CPU usage
    and classifies server load.
    """

    average_cpu = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", round(average_cpu), "%")

    if average_cpu < 50:
        print("Server Status: Normal âœ…")
    elif 50 <= average_cpu <= 80:
        print("Server Status: Warning âš ï¸")
    else:
        print("Server Status: Critical ðŸš¨")


# Sample Input
classify_server_load([45, 60, 70, 85, 90])
```

```python
# Problem 5: Smart Classroom Resource Usage Monitor

def resource_monitor(resource_usage):
    """
    This function tracks resource usage
    and identifies overused resources.
    """

    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:  # Assume 8 hours is threshold
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes âš¡")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(usage)
```

```python
# Problem 6: Online Event Registration Capacity Controller

def event_registration(capacity, registrations):
    """
    This function manages event registrations,
    prevents overbooking, and triggers waitlist.
    """

    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Input
event_registration(100, 105)
```
